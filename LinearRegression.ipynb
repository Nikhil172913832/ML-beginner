```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
ages = np.array([20,25, 30, 35, 40, 45, 50, 55, 60, 65]).reshape((-1,1))
weights = np.array([70, 72, 75, 80, 85, 88, 92, 95, 97, 100])
model = LinearRegression()
model.fit(ages, weights)
m = model.coef_
c = model.intercept_
weight_pred = model.predict(ages)
plt.scatter(ages, weights, color = 'blue')
plt.plot(ages, weight_pred, color='red')
plt.xlabel('Age (years)')
plt.ylabel('Weight (kg)')
plt.title('Linear Regression of Weight on Age')
plt.grid(True)
plt.show()
```
